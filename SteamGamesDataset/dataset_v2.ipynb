```python
# Balatro 게임 검색
balatro_check = df[df['name'] == 'Balatro']

# 결과 출력
if len(balatro_check) > 0:
    print("Balatro 게임이 존재합니다:")
    print(balatro_check.iloc[0])
else:
    print("Balatro 게임이 데이터셋에 없습니다.")
```

```python


# 필터링 전 데이터 수 저장
before_count = len(df)

# name이 알파벳, 숫자, 한글로 시작하는 게임만 유지
df = df[df['name'].str.match('^[a-zA-Z0-9가-힣]')]

# 필터링 후 데이터 수와 제거된 행 수 계산
after_count = len(df)
removed_count = before_count - after_count

print(f"특수문자로 시작하는 게임 제거 전 데이터 수: {before_count}")
print(f"특수문자로 시작하는 게임 제거 후 데이터 수: {after_count}")
print(f"제거된 행 수: {removed_count}")
```
